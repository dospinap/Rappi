```python
#la siguiente funcion recibe 5 parametros:
#La base de datos a sacar la informacion
#la importancia del factor aroma (1 a 5)
#La importancia del factor sabor (1 a 5)
#La importancia del factor apariencia (1 a 5)
#La importancia del factor paleta (1 a 5)
def user_preferences(dataset, aroma, sabor, apariencia, paleta):
    #Se le la basa de datos como un dataframe de pandas
    data = pd.read_csv(dataset, error_bad_lines=False)
    data2 = data
    #Se agrupa la informacion por cerveza, contando la cantidad de usuarios distintos que calificaron cada cerveza
    tmp_df1 = data.groupby(["id_cerveza"]).agg({"usuario": lambda x: x.nunique()})
    
    #Debido a que algunas cervezas tienen muy pocas calificaciones por usuarios distintos se decide filtar la muestra usando
    #la media y la desviacion estandar de cantidad de califiaciones de usuarios distintos
    std, mean = tmp_df1["usuario"].std() , tmp_df1["usuario"].mean()
    tmp_df1.columns = ['total_usuarios_distintos']
    indexes = np.array(tmp_df1.loc[tmp_df1["total_usuarios_distintos"] < (std + mean)].index)    
    data = data.set_index('id_cerveza')
    data = data.drop(indexes)
    
    #Se calculan los pesos de cada parametro de preferencia
    aroma = float(5 if aroma > 5 else aroma)
    aroma = float(1 if aroma < 1 else aroma)
    sabor = float(5 if sabor > 5 else sabor)
    sabor = float(1 if sabor < 1 else sabor)
    apariencia = float(5 if apariencia > 5 else apariencia)
    apariencia = float(1 if apariencia < 1 else apariencia)
    paleta = float(5 if paleta > 5 else paleta)
    paleta = float(1 if paleta < 1 else paleta)
    total = aroma + sabor + apariencia + paleta
    w_aroma = aroma/total
    w_sabor = sabor/total
    w_apariencia = apariencia/total
    w_paleta = paleta/total
    
    #Se calculan las califaciones promedio de cada factor agrupado por cerveza
    tmp_df3 = data.groupby(["id_cerveza"]).agg({"calificacion_total": np.mean})
    tmp_df6 = data.groupby(["id_cerveza"]).agg({"calificacion_aroma": np.mean})
    tmp_df7 = data.groupby(["id_cerveza"]).agg({"calificacion_sabor": np.mean})
    tmp_df8 = data.groupby(["id_cerveza"]).agg({"calificacion_apariencia": np.mean})
    tmp_df9 = data.groupby(["id_cerveza"]).agg({"calificacion_paleta": np.mean})
    tmp_df10 = tmp_df6.join(tmp_df7).join(tmp_df8).join(tmp_df9)
    
    #Se ponderan las calificaciones promedio de acuerdo a los pesos de cada parametro de preferencia
    tmp_df10["user_aroma"] = tmp_df10["calificacion_aroma"]*w_aroma
    tmp_df10["user_sabor"] = tmp_df10["calificacion_sabor"]*w_sabor
    tmp_df10["user_apariencia"] = tmp_df10["calificacion_apariencia"]*w_apariencia
    tmp_df10["user_paleta"] = tmp_df10["calificacion_paleta"]*w_paleta
    tmp_df10["user_total"] = tmp_df10["user_aroma"] + tmp_df10["user_sabor"] + tmp_df10["user_apariencia"] + tmp_df10["user_paleta"]
    
    #Se ordena la base y escoge la cerveza con la ponderacion mas alta
    tmp_df10 = tmp_df10.sort_values("user_total", ascending = False)   
    best_id = tmp_df10.index[0]
    tmp_df12 = data2[data2['id_cerveza'] == best_id]
    best_guess = tmp_df12["nombre_cerveza"].values[0]
    print('Deberias comprar una cerveza "' + best_guess + '"!')
```
